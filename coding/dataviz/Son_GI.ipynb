# Plot Goals involvements according to their Age for Son

## Imports

```python
import pandas as pd 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import matplotlib.patheffects as path_effects
import numpy as np
from highlight_text import fig_text
```

## Retrieve data

```python
df_7 = pd.read_csv("../../data/goal_involments/son_heungmin.csv")
```

```python
df_7
```

```python
df_7['GI']=df_7['Gls']+df_7['Ast']
```

```python
df_7
```

## Create the initial bar plot

```python
fig,ax = plt.subplots(figsize = (11,7))

#Add grid to the plot
ax.grid(ls="dotted", lw="0.5", color="w", zorder=1)

hfont = {'fontname':'Impact'}

#Change the background color
fig.set_facecolor("#08143D")
ax.patch.set_facecolor("#08143D")

#Change the ticks color
ax.tick_params(colors="w", length=15)

#Change the axes color
ax.spines['bottom'].set_color("w")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color("w")

# Plot Son Goal Involments
plt.bar(df_7['Age'], df_7['GI_CumSum'], zorder=1, color="#2bb6e0",edgecolor = 'w')

# Split the graph in Club areas
ax.axvline(x=21, color="w", zorder=0)
ax.axvline(x=23, color="w", zorder=0)
## Hamburger SV
hamburgerpart = ax.text(19, 210, "Hamburger SV", fontsize=16, ha="center", color="w",alpha=0.4 , **hfont, zorder=0)
## Leverkusen 
leverkusenpart = ax.text(22, 210, "Leverkusen", fontsize=16, ha="center", color="w",alpha=0.4 , **hfont, zorder=0)
## Spurs
spurspart = ax.text(25, 210, "Spurs", fontsize=16, ha="center", color="w",alpha=0.4 , **hfont, zorder=0)

# Plot Goal Involmements total at various ages.
## 20
age20 = plt.text(x=20, y=24, s=22, ha="center", size=10, color="w")
#age20 = plt.text(x=20, y=24, s=22, ha="center", size=10, color="#2bb6e0")
#age20.set_path_effects([path_effects.withStroke(linewidth=2, foreground="w")])

# 22
age22 = plt.text(x=22, y=58, s=56, ha="center", size=10, color="w")
#age22 = plt.text(x=22, y=58, s=56, ha="center", size=10, color="#2bb6e0")
#age22.set_path_effects([path_effects.withStroke(linewidth=2, foreground="w")])

## 25
age25 = plt.text(x=25, y=129, s=127, ha="center", size=10, color="w")
#age25 = plt.text(x=25, y=129, s=127, ha="center", size=10, color="#2bb6e0")
#age25.set_path_effects([path_effects.withStroke(linewidth=2, foreground="w")])

# 27
age27 = plt.text(x=27, y=188, s=186, ha="center", size=10, color="w")
#age27 = plt.text(x=27, y=188, s=186, ha="center", size=10, color="#2bb6e0")
#age27.set_path_effects([path_effects.withStroke(linewidth=2, foreground="w")])

# 28
age28 = plt.text(x=28, y=223, s=221, ha="center", size=10, color="w")
#age28 = plt.text(x=28, y=223, s=221, ha="center", size=10, color="#2bb6e0")
#age28.set_path_effects([path_effects.withStroke(linewidth=2, foreground="w")])

# Set the ticks
plt.xticks([18, 20, 22, 24,26, 28])
plt.yticks([0,50,100,150,200,250])
ax.set_ylim(0)
plt.xlim(xmin=17, xmax=29)

# Axes Label
plt.xlabel("Age", color="w", fontsize=15 , **hfont)
plt.ylabel("Total Goal Involvements", color="w", fontsize=15 , **hfont)

# Title and credits
fig_text(0.12,0.97, s="<Son Heung-Min> is more lethal and decisive than ever\n", highlight_textprops=[{"color": '#2bb6e0'}], fontsize = 16, fontweight = "bold", color = "w", **hfont)
fig.text(0.12,0.9, "Consistantly getting better and better", fontsize = 12, fontweight="light", color = "w", **hfont)
fig.text(0.05, 0, "Created by Jacky KUOCH",fontstyle="italic",fontsize=9, color="w")

plt.show()
plt.tight_layout()
```
