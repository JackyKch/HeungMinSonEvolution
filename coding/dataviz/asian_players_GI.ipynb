# Plot Goals involvements according to their Age for best Asian players in history 

## Imports

```python
import pandas as pd 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import matplotlib.patheffects as path_effects
import numpy as np
from highlight_text import fig_text
```

## Retrieve data

```python
df_1 = pd.read_csv("../../data/goal_involments/honda_keisuke.csv")
df_2 = pd.read_csv("../../data/goal_involments/kagawa_shinji.csv")
df_3 = pd.read_csv("../../data/goal_involments/nakamura_shunsuke.csv")
df_4 = pd.read_csv("../../data/goal_involments/nakata_hidetoshi.csv")
df_5 = pd.read_csv("../../data/goal_involments/okazaki_shinji.csv")
df_6 = pd.read_csv("../../data/goal_involments/park_jisung.csv")
df_7 = pd.read_csv("../../data/goal_involments/son_heungmin.csv")
```

## Create the plot

```python
fig,ax = plt.subplots(figsize = (11,7))

#Add grid to the plot
ax.grid(ls="dotted", lw="0.5", color="w", zorder=0)

hfont = {'fontname':'Impact'}

#Change the background color
fig.set_facecolor("#08143D")
ax.patch.set_facecolor("#08143D")

#Change the ticks color
ax.tick_params(colors="w", length=15)

#Change the axes color
ax.spines['bottom'].set_color("w")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color("w")

# Plot Honda Goal Involments
plt.step(df_1['Age'], df_1['GI_CumSum'], where='pre', zorder=1, color="#D200A4")
ax.scatter(33, 133, s=120, color="#D200A4", edgecolors="w", lw=1, zorder=3 )
sonmax = ax.text(33.5, 140, "Keisuke Honda", fontsize=13, ha="center", color="#D200A4", **hfont)
sonmax.set_path_effects([path_effects.withStroke(linewidth=3, foreground="w")])

# Plot Kagawa Goal Involments
plt.step(df_2['Age'], df_2['GI_CumSum'], where='pre', zorder=1, color="#8BF600")
ax.scatter(31, 123, s=120, color="#8BF600", edgecolors="w", lw=1, zorder=3 )
sonmax = ax.text(30, 135, "Shinji\nKagawa", fontsize=13, ha="center", color="#8BF600", **hfont)
sonmax.set_path_effects([path_effects.withStroke(linewidth=3, foreground="w")])

# Plot Nakamura Goal Involments
plt.step(df_3['Age'], df_3['GI_CumSum'], where='pre', zorder=1, color="#FF63FF")
ax.scatter(42, 133, s=120, color="#FF63FF", edgecolors="w", lw=1, zorder=3 )
sonmax = ax.text(41, 143, "Shunsuke\nNakamura", fontsize=13, ha="center", color="#FF63FF", **hfont)
sonmax.set_path_effects([path_effects.withStroke(linewidth=3, foreground="w")])

# Plot Nakata Goal Involments
plt.step(df_4['Age'], df_4['GI_CumSum'], where='pre', zorder=1, color="#03D9E3")
ax.scatter(28, 57, s=120, color="#03D9E3", edgecolors="w", lw=1, zorder=3 )
sonmax = ax.text(29.5, 30, "Hidetoshi\nNakata", fontsize=13, ha="center", color="#03D9E3", **hfont)
sonmax.set_path_effects([path_effects.withStroke(linewidth=3, foreground="w")])

# Plot Okazaki Goal Involments
plt.step(df_5['Age'], df_5['GI_CumSum'], where='pre', zorder=1, color="#A672FF")
ax.scatter(34, 84, s=120, color="#A672FF", edgecolors="w", lw=1, zorder=3 )
sonmax = ax.text(35.5, 71, "Shinji Okazaki ", fontsize=13, ha="center", color="#A672FF", **hfont)
sonmax.set_path_effects([path_effects.withStroke(linewidth=3, foreground="w")])

# Plot Park Goal Involments
plt.step(df_6['Age'], df_6['GI_CumSum'], where='pre', zorder=1, color="#C60070")
ax.scatter(32, 72, s=120, color="#C60070", edgecolors="w", lw=1, zorder=3 )
sonmax = ax.text(33, 59, "Park Ji-Sung", fontsize=13, ha="center", color="#C60070", **hfont)
sonmax.set_path_effects([path_effects.withStroke(linewidth=3, foreground="w")])

# Plot Son Goal Involments
plt.step(df_7['Age'], df_7['GI_CumSum'], where='pre', zorder=1, color="#2bb6e0")
ax.scatter(28, 221, s=120, color="#2bb6e0", edgecolors="w", lw=1, zorder=3 )
sonmax = ax.text(27.9, 230, "Son Heung-Min", fontsize=13, ha="center", color="#2bb6e0", **hfont)
sonmax.set_path_effects([path_effects.withStroke(linewidth=3, foreground="w")])


# Set the ticks
plt.xticks([15, 20, 25,30, 35, 40])
plt.yticks([0,50,100,150,200,250])
ax.set_ylim(0)
plt.xlim(xmin=15, xmax=43)

# Axes Label
plt.xlabel("Age", color="w", fontsize=15 , **hfont)
plt.ylabel("Total Goal Involvements\n (Goals + Assists)", color="w", fontsize=15 , **hfont)

# Title and credits
fig_text(0.12,0.97, s="<Son Heung-Min> could become the best Asian player in history\n", highlight_textprops=[{"color": '#2bb6e0'}], fontsize = 16, fontweight = "bold", color = "w", **hfont)
fig.text(0.12,0.9, "Comparing the best East Asia players", fontsize = 12, fontweight="light", color = "w", **hfont)
fig.text(0.05, 0, "Created by Jacky KUOCH",fontstyle="italic",fontsize=9, color="w")



plt.savefig("test.png")
plt.show()
plt.tight_layout()

```
