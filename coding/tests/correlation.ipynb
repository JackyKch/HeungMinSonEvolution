```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns 
```

```python
df = pd.read_csv("../../data/club/spurs/Son_data_champ_matchs_details_spurs.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
columns = ['Points_Won','Start','Gls','Touches','Min','Home_1_Away_0','SonTeamScore',]
```

```python
dftest = df[columns]
```

```python
dftest.Gls = dftest.Gls.replace("On m", None)
```

```python
dftest['Gls'] = pd.to_numeric(dftest.Gls)
```

```python
dftest.Min = dftest.Min.replace("On m", None)
```

```python
dftest['Min'] = pd.to_numeric(dftest.Min)
```

```python
dftest.Touches = dftest.Touches.replace("On m", None)
```

```python
dftest['Touches'] = pd.to_numeric(dftest.Touches)
```

```python

plt.figure(figsize=(10,10))
sns.heatmap(dftest.corr("spearman"),
            vmin=-1, vmax=1,
            cmap='coolwarm',
            annot=True, 
            square=True)
```
