## Default Matplotlib

```python
import matplotlib as mpl
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(8,8))
fig.set_facecolor("white")

ax.grid(linewidth = 0.25, zorder = 1)

ax.set_title("Title")

np.random.seed(667)
x = np.random.uniform(0,1,50)
y = np.random.uniform(0,1,50)

ax.scatter(x,y, zorder = 2, s = 200, edgecolors= "white", linewidths= 0.5)

plt.tight_layout()

plt.savefig("../coding/default_mpl", bbox_inches = "tight", dpi = 300)
plt.show()
```

## Create a figure

```python
fig, ax = plt.subplots(figsize=(8,8))

fig.set_facecolor('#444444') #figure background dark grey
ax.patch.set_facecolor("grey") #axes background light grey

fig.text(0.05, 1, "This is the figure", fontsize =20) #text on the figure

ax.text (0.5, 0.5, "This is the axes object", ha="center", fontsize =20) #text centered on the axes

plt.tight_layout()
#plt.savefig("../coding/figure", dpi = 300, bbox_inches = "tight")
plt.show()
```

## Lines

```python
fig, ax = plt.subplots(figsize=(8,8))

x = [0,0.1,0.2,0.3,0.4]
y = [0.2,0.8,0.7,0.9,1]

ax.plot(x,y, color="red")
```

```python
#individual lines

fig, ax = plt.subplots(figsize=(8,8))
x = [0,0.1,0.2,0.3,0.4]
y = [0.2,0.8,0.7,0.9,1]
ex = [0.9,0.4,0.3,0.7,0.1]
ey = [0.1,0.3,0.6,0.1,0.9]

for x,y,ex,ey in zip (x,y,ex,ey) :
    ax.plot([x,ex], [y,ey], color = "blue")
```

## Markers

```python
fig, ax = plt.subplots(figsize=(8,8))
x = [0,0.1,0.2,0.3,0.4]
y = [0.2,0.8,0.7,0.9,1]

ax.scatter(x, y, marker="d", s = 200, alpha=0.4, hatch = "///", edgecolors='black')
#s for size
#marker for shape
#alpha for transparency
#hatch for crossing a shape
#edgescolors for edges colors
```

## Ticks and Spines 

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.spines['top'].set_visible(False)

ax.tick_params(axis="x", color="blue", length = 10, width = 2)

ax.tick_params(axis="y", length =0, labelcolor = "blue", labelsize =16, labelrotation = 45)
```

from IPython.core.display import HTML

def make_html (fontname, typetest_string) :

    return "<p><b>{font}:</b> <br/><span style='font-family:{font}; font-size : 20px; '>{typetest_string}</p><br>".format(typetest_string = typetest_string, font =fontname)

typetest_string = "The quick brown fox jumps over the lazy dog"

code = "\n".join([make_html(font, typetest_string) for font in sorted(set([f.name for f in mpl.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count : 2;'>{}</div>".format(code))

## Changing Fonts 

```python
title_font = 'Alegrya Sans'
main_font = 'Open Sans' 

fig, ax = plt.subplots(figsize=(8,8))
fig.set_facecolor("w")
ax.patch.set_facecolor("w")

np.random.seed(667)
x = np.random.uniform(0,1,100)
y = np.random.uniform(0,1,100)

ax.set_title("This is a title using default matplotlib", loc="left", fontweight ="bold", fontsize = 18, fontfamily = title_font)

ax.grid(lw=0.5, color ="k", zorder =1)

ax.scatter(x,y, color ="dodgerblue", edgecolors="w", linewidths=1, s=400, zorder = 2)

ax.text(x[18], y[18]+0.025, "Point Label", ha="center", color="k", fontsize =14, fontweight ="medium", zorder=3, fontfamily = main_font)

ax.set_xlabel("This is an x label", fontweight ="bold", fontfamily = title_font, size =16)
ax.set_ylabel("This is a y label", fontweight = "bold", fontfamily = title_font, size = 16)
ax.set_xlim(0,1)
ax.set_ylim(0,1)

plt.show()
```
